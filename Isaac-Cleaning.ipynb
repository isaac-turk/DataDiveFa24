```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df_21_22_roster = pd.read_csv('2021-2022/21-22-Roster.csv')
df_21_22_roster_pg = pd.read_csv('2021-2022/21-22-Roster-Per-Game.csv')
df_21_22_to_pg = pd.read_csv('2021-2022/21-22-Team&Opp-Per-Game.csv')
df_21_22_results = pd.read_csv('2021-2022/21-22-Sched&Results.csv')
df_21_22_log = pd.read_csv('2021-2022/21-22-Game-Log.csv')
```

```python
# Drop extra index column
df_21_22_roster = df_21_22_roster.drop(columns=['Unnamed: 0'])
df_21_22_roster
```

```python
df_21_22_roster.info()
```

```python
df_21_22_roster=df_21_22_roster.drop(columns='Summary', axis=1)
df_21_22_roster
```

```python
df_21_22_roster_pg = df_21_22_roster_pg.drop(columns=['Unnamed: 0', 'Rk', 'Awards'])
df_21_22_roster_pg
```

```python
df_21_22_roster_pg.dtypes
```

```python
df_21_22_roster_pg.isna().sum()
```

```python
df_roster = pd.merge(df_21_22_roster, df_21_22_roster_pg, on='Player', how='inner')
df_roster = df_roster.drop(columns='Pos_y')
df_roster.rename(columns={'Pos_x': 'Pos'}, inplace=True)
df_roster
```

```python
df_21_22_to_pg = df_21_22_to_pg.drop(columns=['Unnamed: 0', 'Unnamed: 1', 'G', 'MP'])
df_21_22_to_pg
```

```python
df_21_22_to_pg.insert(0, 'Type', ['Team', 'Team-Rank', 'Opponent', 'Opponent-Rank'])
df_21_22_to_pg
```

```python
df_21_22_results = df_21_22_results.drop(columns=['G', 'Unnamed: 0', 'Unnamed: 5', 'OT'])
df_21_22_results['Home Game'] = df_21_22_results['Arena'] == 'State Farm Center'
df_21_22_results = df_21_22_results.drop(columns='Arena')
df_21_22_results.rename(columns={'Unnamed: 9' : 'Result'}, inplace=True)
df_21_22_results
```

```python
streaks = df_21_22_results['Streak'].str.extract(r'([WL]) (\d+)')
df_21_22_results['Win Streak'] = (streaks[0] == 'W') * streaks[1].astype(int)
df_21_22_results['Lose Streak'] = (streaks[0] == 'L') * streaks[1].astype(int)
df_21_22_results['Win Streak'].fillna(0, inplace=True)
df_21_22_results['Lose Streak'].fillna(0, inplace=True)
df_21_22_results['Win Streak'] = df_21_22_results['Win Streak'].astype(int)
df_21_22_results['Lose Streak'] = df_21_22_results['Lose Streak'].astype(int)
df_21_22_results = df_21_22_results.drop(columns='Streak')
df_21_22_results
```

```python
df_21_22_results.insert(4, 'Rank', None)
df_21_22_results['Rank'] = df_21_22_results['Opponent'].str.extract(r'\((\d+)\)')
df_21_22_results['Rank'] = df_21_22_results['Rank'].apply(lambda x: int(x) if pd.notna(x) else None)
df_21_22_results['Opponent'] = df_21_22_results['Opponent'].str.replace(r'\s*\(\d+\)', '', regex=True)
df_21_22_results.insert(5, 'Ranked Match', None)
df_21_22_results['Ranked Match'] = df_21_22_results['Rank'].notna()
df_21_22_results
```

```python
df_21_22_results.dtypes
```

```python
df_21_22_log = df_21_22_log.drop(columns=['G', 'Unnamed: 0', 'Unnamed: 3'])
df_21_22_log
```

```python
df_merged = pd.merge(df_21_22_results, df_21_22_log, left_index=True,right_index=True, how='inner')
df_merged = df_merged.drop(columns=['W/L', 'Date_x', 'Opp_y', 'Tm_y', 'Opp.1', 'Unnamed: 24'])
df_merged
```

```python
date_column = df_merged.pop('Date_y')
df_merged.insert(0, 'Date', date_column)
df_merged
```

```python
df_merged.rename(columns={'Tm_x' : 'PF', 'Opp_x' : 'PA'}, inplace=True)
df_merged
```

```python
df_merged.columns = [f'OPP {col[:-2]}' if '.1' in col else col for col in df_merged.columns]
df_merged
```

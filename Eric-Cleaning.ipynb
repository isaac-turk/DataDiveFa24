```python
results = results.drop(columns=['G', 'Unnamed: 0', 'Unnamed: 5', 'OT'])
results['Home Game'] = results['Arena'] == 'State Farm Center'
results = results.drop(columns='Arena')
results.rename(columns={'Unnamed: 9' : 'Result'}, inplace=True)
results
```

```python
streaks = results['Streak'].str.extract(r'([WL]) (\d+)')

# Ensure NaNs are replaced with 0 before converting to int
streaks[1] = streaks[1].fillna(0).astype(int)

# Create Win Streak and Lose Streak columns
results['Win Streak'] = (streaks[0] == 'W') * streaks[1]
results['Lose Streak'] = (streaks[0] == 'L') * streaks[1]

# Drop the 'Streak' column safely
results = results.drop(columns='Streak', errors='ignore')

# Print the results to verify
results
```

```python
results.insert(4, 'Rank', None)
results['Rank']=results['Opponent'].str.extract(r'\((\d_)\)')
results['Rank']=results['Rank'].apply(lambda x: int(x) if pd.notna(x) else None)
results['Opponent']=results['Opponent'].str.replace(r'\s*\(\d+\)', '', regex=True)
results.insert(5, 'Ranked Match', None)
results['Ranked Match']=results['Rank'].notna()
results
```

```python
log = log.drop(columns=['G', 'Unnamed: 0', 'Unnamed: 3'])
log
```

```python
df_log = pd.merge(results, log, left_index=True,right_index=True, how='inner')
df_log = df_log.drop(columns=['W/L', 'Date_x', 'Opp_y', 'Tm_y', 'Opp.1', 'Unnamed: 24'])
df_log
```

```python
date_column = df_log.pop('Date_y')
df_log.insert(0, 'Date', date_column)
df_log
```

```python
df_log.rename(columns={'Tm_x' : 'PF', 'Opp_x' : 'PA'}, inplace=True)
df_log
```

```python
df_log.columns = [f'OPP {col[:-2]}' if '.1' in col else col for col in df_log.columns]
df_log
```
